```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = pd.DataFrame({
    'time': range(10),
    'value': [10, 15, 13, 18, 16, 20, 17, 22, 19, 24],
    'category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C']
})

# Create faceted line plot
sns.set(style="ticks")
sns.lineplot(x="time", y="value", hue="category", data=data, markers=True)

plt.show()
```
